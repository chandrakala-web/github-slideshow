# **DEALING** **WITH MISSING VALUES**

**Check if our data contains any missing values**

```python
perc_missing_data = pd.DataFrame([df.isnull().sum(),df.isnull().sum()*100.0/df.shape[0]]).T
perc_missing_data.columns = ['No. of Missing Data', '% Missing Data']
perc_missing_data
```

Now that we have understood that there are 4 features with null values (Company, agent, country, children). Weâ€™ll deal each column separately. Now when we have dig deeper into the dataset to understand about each feature.


Children column contains the count of children, so we will replace all the missing values with the rounded mean value.

And our country column contains country codes representing different countries. It is a categorical feature so I will also replace it with the mode value. The mode value is the value that appears more than any other value. So, in this case, I am replacing it with the country that appears the most often.

```python
# for missing children value, replace it with rounded mean value
df['children'].fillna(round(df.children.mean()), inplace=True)
```

```python
## For the missing values in the country column, replace it with mode (value that appears most often)
df['country'].fillna(df.country.mode().to_string(), inplace=True)
```

```python
## For the missing values in agent column replace it with zero
df['agent'] = df['agent'].fillna(0.0)
```

 Country is having highest number of missing values morem than  45% . So now here we are dropping this column

```python
#Drop the columns agent and company because these data will not require for analysis part.
df =df.drop(columns=['company'], axis = 1)
df.shape
```

Dropping the unnecessary data can imply a reduction of the sample size. This can help us from proceeding with the analysis.



```python
#Removing Duplicates rows
Duplicate_rows_df = df[df.duplicated()]
df = df.drop_duplicates()
```

There are many rows that have zero guests including adults, children and babies. These type of rows does not make.So we will just remove those rows.

```python
## Drop Rows where there is no adult, baby and child
df = df.drop(df[(df.adults+df.babies+df.children)==0].index)
```

# **Checking for outliers**

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes', 'adr','required_car_parking_spaces']
n = 1
plt.figure(figsize=(50,30))

for column in columns:
  plt.subplot(4,4,n)
  n = n+1
  sns.boxplot(df[column])  
```

Yes..we heve outliers lets remove those outliers

```python
df.loc[df.lead_time > 500, 'lead_time'] = 500
df.loc[df.stays_in_weekend_nights >=  5, 'stays_in_weekend_nights'] = 5
df.loc[df.adults > 4, 'adults'] = 4
df.loc[df.previous_bookings_not_canceled > 0, 'previous_bookings_not_canceled'] = 1
df.loc[df.previous_cancellations > 0, 'previous_cancellations'] = 1
df.loc[df.stays_in_week_nights > 10, 'stays_in_week_nights'] = 10
df.loc[df.booking_changes > 5, 'booking_changes'] = 5
df.loc[df.babies > 8, 'babies'] = 0
df.loc[df.required_car_parking_spaces > 5, 'required_car_parking_spaces'] = 0
df.loc[df.children > 8, 'children'] = 0
df.loc[df.adr > 1000, 'adr'] = 1000
```

## **4.To predict whether or not a hotel was likely to receive a disproportionately high number of special requests?**

```python
list_of_special_request = list(df['total_of_special_requests'])
total_number = len(list_of_special_request) #to get total number of data of total special request
```

```python
likely_to_get_request = [] #the data of special request was there.
alikely_to_get_request = []# the data of special request was not there.
for number in list_of_special_request:
  if number == 0:
    alikely_to_get_request.append(number) 
  else:
    likely_to_get_request.append(number)
```

```python
x= len(likely_to_get_request)
hotel_likely_to_get_special_request = (x / total_number)*100
print(f" The hotel was likely to recieve a disproportionality high number off special request is {hotel_likely_to_get_special_request}")
```

# **5.From which market segment most Bookings were received and how it will affect the price per person ?**

```python
segments=df["market_segment"].value_counts()
#pie chart
fig = px.pie(segments,values=segments.values,names=segments.index,title="Bookings per market segment",template="seaborn")
fig.update_traces(rotation=-90, textinfo="percent+label")
fig.show()
```



*   Price per night (ADR)



Since no currency information is given, but Portugal is part of the European Monetary Union, We assume that all prices are in EUR

```python
#Resort data
resort = df.loc[(df["hotel"] == "Resort Hotel")]
#City hotel data
hotel = df.loc[(df["hotel"] == "City Hotel")]
#price column
df["price"] = df["adr"] / (df["adults"] + df["children"])
#count adult and children as paying guest only
resort['adr_price'] = resort["adr"] / (resort["adults"] + resort["children"])
hotel['adr_price'] = hotel["adr"] / (hotel["adults"] + hotel["children"])
```

```python
print("""From all non-canceled bookings, the average prices are:
Resort hotel: {:.2f} â‚¬ per night and person.
City hotel: {:.2f} â‚¬ per night and person.""".format(resort["adr_price"].mean(), hotel["adr_price"].mean()))
```
